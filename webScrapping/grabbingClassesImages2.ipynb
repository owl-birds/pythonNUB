```python
result = requests.get("https://en.wikipedia.org/wiki/Philosophy")
```

```python
print(result)
print(result.text)
```

```python
soup = bs4.BeautifulSoup(result.text,"lxml")
print(soup)
```

```python
toc = soup.select(".toc")
print(toc)
```

```python
for element in toc:
    print(element.getText())
```

```python
tocText = soup.select(".toctext")
for element in tocText:
    print(element.getText())
```

# GRABBING AN IMAGE

```python
pageImages = soup.select("img")
```

```python
print(pageImages)
```

```python
for image in pageImages:
    print(image["src"])
```

```python
thumbImages = soup.select(".thumbimage")
```

```python
print(thumbImages)
```

```python
schoolOfAthens = soup.select(".thumbimage")[0]
print(schoolOfAthens["src"])
```
